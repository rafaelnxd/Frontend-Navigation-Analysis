```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
print(data.describe())
```

```python
# Click Scatter

plt.scatter(data['x_position'], data['y_position'])
plt.title('Click Scatter')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.show()
```

```python
# Time spent on each page

pages = data['page'].unique()

for page in pages:
    page_data = data[data['page'] == page]
    if not page_data.empty: 
        plt.figure()
        sns.histplot(page_data['time_spent'], kde=True)
        plt.title(f'Time spent on: {page}')
        plt.xlabel('Time (ms)')
        plt.ylabel('Frequency')

plt.show()
```
