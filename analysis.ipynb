```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
print(data.describe())
```

```python
# Click Scatter

plt.scatter(data['x_position'], data['y_position'])
plt.title('Click Scatter')
plt.xlabel('Posição X')
plt.ylabel('Posição Y')
plt.show()
```

```python
# Time spent on each page

sns.histplot(data['time_spent'], kde=True)
plt.title('Time spent on each page')
plt.xlabel('ms')
plt.ylabel('Frequency')

```
